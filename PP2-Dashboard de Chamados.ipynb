```python
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
# leitura e limpeza do csv e conversão para DataFrame
df = pd.read_csv('/content/relatorio_chamados.csv',sep=',')

df['Equipe Destinada'] = df['Equipe Destinada'].fillna('Não Informada')
df['Tempo_'] = pd.to_timedelta(df['Tempo de Atendimento'], errors='coerce')
df['Minutos'] = df['Tempo_'].dt.total_seconds() / 60
df = df.drop(columns=['Tempo_'])

# criando menu de seleção de equipe
evc = df['Equipe Destinada'].value_counts()

equipes = evc.index
count = evc.values

print("<RELATÓRIO DE CHAMADOS>")
print("SELECIONE A EQUIPE:")
for i,n in enumerate(equipes):
  print(f'{i} - {n}')

try:
  equipe = input()
  equipe = (int)(equipe)
except ValueError:
    print("Erro: digite um número válido.")
    exit()

if(equipe < 0 or equipe >= len(equipes)):
  print('Erro: equipe selecionada não se encontra no relatório.')
  exit()

equipe = equipes[equipe]
df_filter = df[df['Equipe Destinada'] == equipe]

# criação da visualização compartilhada
fig, axs = plt.subplots(2,2,constrained_layout=True)

# gráfico de barras
color = ['#de8264' if i == equipe else '#70c4e0' for i in equipes]
axs[0,0].barh(equipes,count,color=color)
axs[0,0].set_title('Quantidade de Chamados', fontsize=6)
axs[0,0].tick_params(axis='x', rotation=45)

# gráfico de pizza
svc = df_filter['Status'].value_counts()

axs[0,1].pie(svc.values,labels=svc.values,startangle=90,colors =['#7be096','#d17979'],labeldistance=0.4,textprops=dict(color="w")
)
axs[0,1].set_title(f'Resolvidos x Pendentes - {equipe}', fontsize=6)
axs[0,1].legend(svc.index,title='Status dos Chamados:',loc='upper left',bbox_to_anchor=(0.5, 0.4))

# histograma
df_filter = df_filter.dropna()

axs[1,0].hist(df_filter['Minutos'],bins='auto',alpha=0.7,edgecolor='black')
axs[1,0].set_title(f'Tempo de Atendimento - {equipe}', fontsize=6)
axs[1,0].set_xlabel('Tempo de Atendimento (min)')

# boxplot
axs[1,1].boxplot(df_filter['Minutos'], vert=False, showmeans=True, meanline=True)
axs[1,1].set_title(f'Distribuição do Tempo de Atendimento - {equipe}', fontsize=6)

plt.tight_layout()
plt.show()
```
